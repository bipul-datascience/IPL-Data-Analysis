```python
#Extracting the records where a team won batting first
batting_first=matches[matches['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
#Making a histogram 
plt.figure(figsize=(7,7))
plt.hist(batting_first['win_by_runs'])
plt.title("Distribution of Runs")
plt.xlabel("Runs")
plt.show()
```

```python
#Finding out the number of wins w.r.t each team after batting first
batting_first['winner'].value_counts()
```

```python
#Making a bar-plot for top 5 teams with most wins after batting first
plt.figure(figsize=(12,6))
plt.bar(list(batting_first['winner'].value_counts().head().keys()),list(batting_first['winner'].value_counts().head()),color=["blue","yellow","orange","purple","red"])
plt.show()
```

```python
#Making a pie chart
plt.figure(figsize=(10,10))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
